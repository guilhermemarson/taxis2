```python
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
import os
path = './src/'
arquivos = os.listdir(path)
```

```python
contador=1
for f in arquivos:
    df = pd.read_csv(path+f)
    geolocator = Nominatim()
    print('Iniciando arquivo: '+ f)
    for i in range(df.shape[0]):     
        try:
            location = geolocator.reverse(str(df['latitude'][i]) +','+ str(df['longitude'][i]) ,  timeout=15) 
            df.ix[i, 'road'] = location.raw['address']['road']
            if i%1000==0:
                print(i)    
        except:
            geolocator = Nominatim()
            print('deu ruim na linha ' +str(i)+' do arquivo '+ f)
    df.to_csv('./output/taxi'+str(contador)+'.csv')
    contador+=1
```
